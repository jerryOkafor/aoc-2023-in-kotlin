# Day 9: Mirage Maintenance
## Part 1

```kotlin
val input = readInputForNotebook("Day09")

input
```

```kotlin
fun List<List<Long>>.printPretty() {
    val list = this
    buildString {
        for (i in 0 until list.size) {
            repeat(i) { append(" ") }
            for (j in 0 until list[i].size) {
                append(" ${list[i][j]} ")
            }
            append("\n")
        }
    }.println("History interporation")
}

fun List<Long>.interporate() = this.windowed(2).map { (num1, num2) ->
    num2 - num1
}

val part1 = input.sumOf { line ->
    val sequences = mutableListOf(line.split("\\s++".toRegex()).map { it.trim().toLong() })

    while (sequences.last().any { it != 0L })
        sequences += sequences.last().windowed(2).map { (num1, num2) -> num2 - num1 }

    sequences.sumOf { it.last() }
}

part1
```

## Part 2

```kotlin
val part2 = input.map { line ->
    val sequences = mutableListOf(line.split("\\s++".toRegex()).map { it.trim().toLong() })

    while (sequences.last().any { it != 0L })
        sequences += sequences.last().windowed(2).map { (num1, num2) -> num2 - num1 }

    sequences
}

part2.joinToString{"${it.printPretty()} \n"}.println()

val result = part2.map { it.map { it.first() } }
result.println()

val result2 = result.map { it.reversed().reduce{acc, l -> l  - acc } }
result2.println()

result2.sum()
```
